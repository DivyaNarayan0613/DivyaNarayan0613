```python
import python_avatars as pa

my_avatar = pa.Avatar(
    style=pa.AvatarStyle.TRANSPARENT,
    background_color=pa.BackgroundColor.BLACK,
    top=pa.HairType.ASTRONAUT,
    eyebrows=pa.EyebrowType.RAISED_EXCITED,
    eyes=pa.EyeType.SQUINT,
    nose=pa.NoseType.DEFAULT,
    mouth=pa.MouthType.SMILE,
    facial_hair=pa.FacialHairType.BEARD_MEDIUM,
    # You can use hex colors on any color attribute...
    skin_color=pa.SkinColor.YELLOW,
    # Or you can use the colors provided by the library
    hair_color=pa.HairColor.BLACK,
    accessory=pa.AccessoryType.SUNGLASSES,
    clothing=pa.ClothingType.SHIRT_V_NECK,
    clothing_color=pa.ClothingColor.PASTEL_GREEN
)

# Save to a file
my_avatar.render("my_avatar.svg")
```
